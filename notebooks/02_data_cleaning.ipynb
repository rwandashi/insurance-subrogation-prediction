# Deleting columns that will not be used to train the model

```python
# Delete one column ( Claim number)
df = df.drop("claim_number", axis=1)
```

# Changing Column datatypes

```python
# Changing columns types
df["subrogation"] = df["subrogation"].astype(bool)
```

```python
df["zip_code"] = df["zip_code"].astype("string")
```

```python
df["claim_date"] = pd.to_datetime(df["claim_date"])
```

```python
df.info()
```

# Transforming floats to boolean

```python
df[["email_or_tel_available", "high_education_ind", "address_change_ind", "policy_report_filed_ind"]].head()
```

```python
# Transforming binary columns from float to boolean
# ["email_or_tel_available", "high_education_ind", "address_change_ind", "policy_report_filed_ind"]
df["email_or_tel_available"] = df["email_or_tel_available"].astype(bool)
df["high_education_ind"] = df["high_education_ind"].astype(bool)
df["address_change_ind"] = df["address_change_ind"].astype(bool)
df["policy_report_filed_ind"] = df["policy_report_filed_ind"].astype(bool)
```

```python
df[["email_or_tel_available", "high_education_ind", "address_change_ind", "policy_report_filed_ind"]].head()
```

# Handling the Yes/No columns

```python
df[['witness_present_ind', 'in_network_bodyshop']].head()
```

```python
# Transforming yes/no questions to 1/0
# ['witness_present_ind', 'in_network_bodyshop']
# Convert Y/N and yes/no to boolean True/False
df["witness_present_ind"] = df["witness_present_ind"].str.upper().map({'Y': True, 'N': False})
df["in_network_bodyshop"] = df["in_network_bodyshop"].str.lower().map({'yes': True, 'no': False})
df[['witness_present_ind', 'in_network_bodyshop']].head()
```

```python
df.info()
```

# Handing the Zip code column

```python
# Delete one column ( zip_code)
df = df.drop("zip_code", axis=1)
```
